### To remove / filter out unnecessary columns & create new dataframe OR subset of original dataframe

```python
tree_census_subset = tree_census[['tree_id', 'tree_dbh', 'stump_diam',
       'curb_loc', 'status', 'health', 'spc_latin', 'steward',
       'sidewalk', 'problems', 'root_stone', 'root_grate', 'root_other',
        'trunk_wire', 'trnk_light', 'trnk_other', 'brch_light', 'brch_shoe', 'brch_other']]


tree_census_subset
# type(tree_census_subset)
```

### To check if there is missing values(i.e NaN ) in the dataframe using  `isna().sum()`

```python
# NOTE: isna().sum() gare pachi matra tha huncha kun column ma total kati ota NaN values cha vanera

# NOTE: isna() le None lai NaN jasto missing values vandaina


tree_census_subset.isna().sum()
```

### Boolean Mask
####  Filtering rows in dataframe by creating a boolean filtering mask (i.e conditional filtering mask)

#### Masking is an advanced concept in Pandas where the analyst tries to filter data based on a particular condition.

```python
# Stump  OR  Dead trees ko rows lai filter garna boolean mask banako
mask = ((tree_census_subset['status'] =='Stump') | (tree_census_subset['status'] =='Dead'))

mask
```

### syntax of loc[  ]:   `loc[row_indexer, column_indexer]`      OR     `loc[row_indexer]`

### using `fillna()` function to fill  `NaN` values

```python
# NOTE:  .loc[mask]  is like .loc[row_indexer]

tree_census_subset.loc[mask] = tree_census_subset.loc[mask].fillna('Not Applicable')
```

### Replacing most tricky `NaN` values from dataframe,
### by, most frequent occuring values in that partiular column

```python
tree_census_subset['problems'].fillna('None', inplace=True)
tree_census_subset['steward'].fillna('1or2', inplace=True)
tree_census_subset['health'].fillna('Good', inplace=True)
tree_census_subset['spc_latin'].fillna('No Observation', inplace=True)
tree_census_subset['sidewalk'].fillna('NoDamage', inplace=True)
```
