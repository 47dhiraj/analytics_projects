###  Visualizing `World Population Percentage` of `most populated countries`

```python
# applying filtering condition and creting population_countries subset dataframe
populated_countries = df[df['World Population Percentage'] > 1.50 ]

populated_countries[['Country', 'World Population Percentage']].plot(kind='scatter', x='Country', y='World Population Percentage', figsize=(15, 5))
```

### For checking/viewing the Correlation between the data

#### correlation between only numeric columns can be found (i.e float64 & int64)

```python
df_numeric = df.select_dtypes(include=['float64', 'int64'])

# You can provide different methods to find correlation: pearson, kendall, spearman OR even pass a custom callable function.
corr = df_numeric.corr(method='pearson')

corr
```

### visualize these correlations using `seaborn` `heatmap`

```python
sns.heatmap(df_numeric.corr(), annot=True)   # For generating the heatmap
# annot=True adds annotations to each cell of the heatmap, displaying the numeric values of the correlation coefficients.

plt.rcParams['figure.figsize'] = (15, 5)      # For changing the size of the figure/plot
plt.show()
```

### Grouping the data using `groupby()`

```python
# df2 = df.groupby('Continent').mean(numeric_only = True).sort_values(by="2022 Population",ascending=False)

# df2
```

```python
# df2.plot()
```

```python
df2 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean(numeric_only = True).sort_values(by="2022 Population", ascending=False)

df2
```

```python
# Transposing the Dataframe
df3 = df2.transpose()

df3
```

```python
df3.plot()
```
